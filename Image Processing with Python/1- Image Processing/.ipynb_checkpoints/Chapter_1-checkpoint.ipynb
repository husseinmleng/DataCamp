# Images with Sckit-Image 

```python
import numpy as np
import matplotlib.pyplot as plt
import skimage 
from skimage.io import imread
from skimage.color import rgb2gray
```

```python
def show_image(image, title='Image', cmap_type='gray'):
    plt.clf()
    plt.figure()
    plt.imshow(image, cmap= cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def show_plot(fig,ax):
    pass
```

```python
# Import the modules from skimage
from skimage import data, color

# Load the rocket image
rocket = data.rocket()

# Convert the image to grayscale
gray_scaled_rocket = color.rgb2gray(rocket)

# Show the original image
show_image(rocket, 'Original RGB image')

# Show the grayscale image
show_image(gray_scaled_rocket, 'Grayscale image')
```

```python
madrid_image = plt.imread('images/madrid.jpg')
vertically_flipped = np.flipud(madrid_image)
show_image(vertically_flipped,'Vertically flipped image')
```

```python
# Obtain the red channel
red_channel = madrid_image[:, :, 0]

# Plot the red histogram with bins in a range of 256
plt.hist(red_channel.ravel(), bins=256)

# Set title and show
plt.title('Red Histogram')
plt.show()
```

# Thresholding

```python
image = imread('images/Cameraman.png',as_gray=True)
image.shape
```

```python
# obtain threshold value
thresh = 127

#apply thresholding to the image 
binary = image > thresh
# Show the original and thresholded
show_image(image, 'Original');
show_image(binary, 'Thresholded');
```

```python
# Apply thresholding to the image
inverted_binary = image <= thresh
# Show the original and thresholded
show_image(image, 'Original')
show_image(inverted_binary,'Inverted thresholded')
```

### Try more thresholding algorithms

```python
from skimage.filters import try_all_threshold

#obtain all the resulting images 
fig, ax = try_all_threshold(image)

# Showing resulting plots
plt.show()
```

### Optimal thresh value (global)

```python
# import the otsu threshold function
from skimage.filters import threshold_otsu

#obtain the optimal threshold value
thresh = threshold_otsu(image)

#apply thresholding to the image 
binary_global = (image > thresh).astype('float')
show_image(image,'Orignal')
show_image(binary_global,'Global Thresholding')
```

### Optimal thresh value (lcoal)

```python
# import local threshold function
from skimage.filters import threshold_local

#set the block size to 35
block_size = 35

#obtain the optimal thresholding value
local_thresh = threshold_local(image,block_size,offset=10)

# Apply local thrsholding  to obtain binary image
binary_local = (image > local_thresh).astype('float64')
# Show the original and binarized image
show_image(image, 'Original')
show_image(binary_local, 'Local thresholding')
```
